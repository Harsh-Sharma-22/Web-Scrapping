### Beautiful Soup and requests

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_Indian_states_and_union_territories_by_GDP_per_capita'
```

```python
page=requests.get(url)
```

```python
soup=BeautifulSoup(page.text,'html')
```

```python
print(soup.prettify())
```

### Find and Find_all

```python
table=soup.find_all('table')[1]
```

```python
print(table)
```

```python
state_titles=table.find_all('th')
```

```python
state_titles
```

```python
state_table_titles=[title.text.strip() for title in state_titles]
print(state_table_titles)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns=state_table_titles)
```

```python
df
```

```python
column_data=table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data=row.find_all('td')
    individual_row_data=[data.text.strip() for data in row_data]
    print(individual_row_data)
    
    length=len(df)
    df.loc[length]=individual_row_data
```
